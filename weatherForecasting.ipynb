```python
categorical_columns = []
numerical_columns = []

# Separate columns based on data type
for col in data.columns:
    if data[col].dtype == 'object':
        categorical_columns.append(col)
    elif data[col].dtype == 'float64' or data[col].dtype == 'int32':
        numerical_columns.append(col)
print(categorical_columns)
print(numerical_columns)
```

```python

# Function to remove outliers by capping
def remove_outliers(df, columns):
    for col in columns:
      print(col)
      q1=data[col].quantile(0.25)
      q3=data[col].quantile(0.75)
      iqr=q3-q1
      ub=q3+(1.5*iqr)
      lb=q1-(1.5*iqr)
      data.loc[(df[col]>ub),col]=ub
      data.loc[(df[col]<lb),col]=lb
    return df



# Apply capping to remove outliers
data = remove_outliers(data, numerical_columns)



```

```python

from sklearn.preprocessing import LabelEncoder, StandardScaler

# Assuming 'data' is your DataFrame and it contains both object (categorical) type and int (numerical) type columns

# Example lists of column names
# categorical_columns = ['cat_col1', 'cat_col2']
# numerical_columns = ['num_col1', 'num_col2']

print("Categorical columns:", categorical_columns)

# Encoding categorical variables and standardizing numerical variables
encoding_mapping = {}  # Dictionary to store encoding mapping

for col in categorical_columns:
    # Convert column to string to ensure uniform data type
    data[col] = data[col].astype(str)

    label_encoder = LabelEncoder()
    data[col] = label_encoder.fit_transform(data[col])
    encoding_mapping[col] = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))

if numerical_columns:  # Check if numerical_columns list is not empty
    scaler = StandardScaler()
    data[numerical_columns] = scaler.fit_transform(data[numerical_columns])

# Print encoding mapping for object type columns
print("\nEncoding mapping:")
for col, mapping in encoding_mapping.items():
    print(f"{col}: {mapping}")
```

```python
def val(column_name, target_key):
    # Accessing the encoding mapping for the specified column
    return encoding_mapping[column_name].get(target_key)


```

```python


# Accessing values in encoding_mapping dictionary
# Let's say you want to access the encoding mapping for the 'Categorical1' column

## Accessing the encoding for 'Agra' in the 'location_name' column
def pred(target_value):
  reversed_weather_mapping = {v: k for k, v in encoding_mapping['Daily Summary'].items()}
  return reversed_weather_mapping.get(target_value)


```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score

# Load your dataset
# Assuming your data is stored in a DataFrame called 'data'

# For Wind Speed prediction
wind_speed_data = data[['Precip Type','Temperature (C)', 'Humidity', 'Visibility (km)', 'Pressure (millibars)','Daily Summary','Wind Direction', 'year', 'month', 'day', 'hour', 'minute','second','day_of_week']]
wind_speed_target = data['Wind Speed (km/h)']

# For Visibility prediction
visibility_data = data[['Precip Type', 'Temperature (C)', 'Humidity', 'Wind Speed (km/h)','Wind Direction', 'Pressure (millibars)', 'Daily Summary', 'year', 'month', 'day', 'hour', 'minute','second','day_of_week']]
visibility_target = data['Visibility (km)']

# Split data into training and testing sets for Wind Speed
X_wind_train, X_wind_test, y_wind_train, y_wind_test = train_test_split(wind_speed_data, wind_speed_target, test_size=0.2, random_state=42)

# Split data into training and testing sets for Visibility
X_vis_train, X_vis_test, y_vis_train, y_vis_test = train_test_split(visibility_data, visibility_target, test_size=0.2, random_state=42)

# Define and train the Random Forest model for Wind Speed
rf_wind_speed_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_wind_speed_model.fit(X_wind_train, y_wind_train)

# Define and train the Random Forest model for Visibility
rf_visibility_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_visibility_model.fit(X_vis_train, y_vis_train)

# Function to evaluate model and print accuracy
def evaluate_model(model, X_test, y_test, target_name):
    predictions = model.predict(X_test)
    mse = mean_squared_error(y_test, predictions)
    r2 = r2_score(y_test, predictions)
    print(f"Mean Squared Error for {target_name}:", mse)
    print(f"R-squared for {target_name}:", r2)
    return mse, r2

# Evaluate Random Forest model for Wind Speed
print("\nRandom Forest for Wind Speed:")
rf_wind_speed_mse, rf_wind_speed_r2 = evaluate_model(rf_wind_speed_model, X_wind_test, y_wind_test, "Wind Speed")

# Evaluate Random Forest model for Visibility
print("\nRandom Forest for Visibility:")
rf_visibility_mse, rf_visibility_r2 = evaluate_model(rf_visibility_model, X_vis_test, y_vis_test, "Visibility")
```
